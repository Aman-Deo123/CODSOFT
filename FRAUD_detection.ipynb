 Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score
```

Loading the Dataset

```python
credit_card_data=pd.read_csv("creditcard.csv")
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# DataSet Informations
credit_card_data.info()
```

```python
 # Checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# Distribution of legit transactions  and fraud transactions 
credit_card_data['Class'].value_counts()
```

```python
# Seperating the data for analysis
legit= credit_card_data[credit_card_data.Class==0]
fraud= credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both transaction
credit_card_data.groupby('Class').mean()
```

```python
# Building a sample dataset containing similar distribution of normal transactions and fraud transactions
# Count of fraud transactions --492
```

```python
legit_sample=legit.sample(n=492)
```

```python
# Concatenating two dataframes
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0) 
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# Splitting the data into Features and Targets
```

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Split the data into Training data and testing Data
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# Model training
```

```python
# LOGISTIC REGRESSION
```

```python
model=LogisticRegression(max_iter=9999)
```

```python
model.fit(X_train,Y_train)
```

```python
# Model  Evaluation 
# Acuuracy Score 
```

```python
# --> Accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print(training_data_accuracy)
```

```python
#--> Accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print(test_data_accuracy)
```
